```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.close('all')
```

```python
titanic = pd.read_csv('./data/titanic.csv')
titanic.head()
```

```python
# Plot the Pclass and the class fare
titanic.groupby('Pclass')['Fare'].count().plot(kind='line');
```

```python
# Horizontal bar graph of female & male where they mounted from and their ages
titanic.groupby(['Embarked','Sex'])['Age'].count().plot(kind='barh', xlabel='Traveler Mounted From/Gender');
```

```python
# Plot number of Female and male who survived age-wise
titanic.groupby(['Survived', 'Sex'])['Age'].count().plot(kind='bar', xlabel='Gender/Survived', ylabel='Number of People');
```

```python
# Scatter plot on PassengerId and Age
titanic.plot(kind='scatter', x='PassengerId', y='Age', color='DarkGreen');
```

```python
# Plot histogram on Age with 5 bins
xticks = np.arange(0,81,5)
yticks = np.arange(0,351,25)

titanic['Age'].plot(kind='hist', bins=5, xticks=xticks, yticks=yticks, rot=45);
```

```python
# Box plot on PassengerId and Fare
color = {
    "boxes": "DarkGreen",
    "whiskers": "DarkOrange",
    "medians": "DarkBlue",
    "caps": "Gray",
}

titanic[['PassengerId','Fare']].plot(kind='box', color=color);
```

```python
# Plot kde on Fare and Age
titanic[['Fare', 'Age']].plot(kind='kde');
```

```python
# Area plot on PassengerId, Fare, and Age
titanic[['PassengerId', 'Fare', 'Age']].plot(kind='area', alpha=0.5, stacked=False)
```

```python
# Hexagonal bin plot on PassengerId and Age
titanic.plot(kind='hexbin', x='PassengerId', y='Age', gridsize=25);
```

```python
# Pie plot on gender percentage
gender_pct = titanic['Sex'].value_counts(normalize=True) * 100
gender_pct.plot(kind='pie', subplots=True, figsize=(8,4));
```
